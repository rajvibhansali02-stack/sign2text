```python
import numpy as np
import os
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

DATA_DIR = "processed_data"

X = []
y = []

for label in os.listdir(DATA_DIR):
    label_path = os.path.join(DATA_DIR, label)
    for file in os.listdir(label_path):
        data = np.load(os.path.join(label_path, file))
        X.append(data)
        y.append(label)

X = np.array(X)

le = LabelEncoder()
y = le.fit_transform(y)

model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

model.fit(X, y)

os.makedirs("model", exist_ok=True)
joblib.dump(model, "model/gesture_model.pkl")
joblib.dump(le, "model/labels.pkl")

print("✅ Model trained and saved successfully")
```

```python
import os
import numpy as np
import joblib
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


DATA_DIR = "processed_data"
MODEL_PATH = "model/gesture_model.pkl"
LABELS_PATH = "model/labels.pkl"

# LOAD MODEL & LABELS
model = joblib.load(MODEL_PATH)
le = joblib.load(LABELS_PATH)

X = []
y_true = []

# LOAD DATA
for label in os.listdir(DATA_DIR):
    label_path = os.path.join(DATA_DIR, label)
    if not os.path.isdir(label_path):
        continue

    for file in os.listdir(label_path):
        if file.endswith(".npy"):
            arr = np.load(os.path.join(label_path, file))
            if arr.shape[0] == 63:
                X.append(arr)
                y_true.append(label)

X = np.array(X)
y_true = np.array(y_true)

print("Total samples:", len(X))

# ENCODE LABELS
y_encoded = le.transform(y_true)

# PREDICT
y_pred_encoded = model.predict(X)
y_pred = le.inverse_transform(y_pred_encoded)

# METRICS
acc = accuracy_score(y_true, y_pred)
print(f"\n✅ Accuracy: {acc*100:.2f}%\n")

print("📋 Classification Report:")
print(classification_report(y_true, y_pred))

print("🧮 Confusion Matrix:")
print(confusion_matrix(y_true, y_pred))
```
